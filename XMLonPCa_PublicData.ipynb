# **RF with 80:20 Split**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)
X_train = pd.DataFrame(X_train)
X_test = pd.DataFrame(X_test)
```

```python
# Assuming X and y are already defined
# Split the data into 80% training and 20% testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Initialize the RandomForestClassifier
clf = RandomForestClassifier(random_state=0)

# Fit the model on the training data
clf.fit(X_train, y_train)
```

```python
# Make predictions on the test data
y_pred = clf.predict(X_test)
y_pred_proba = clf.predict_proba(X_test)[:, 1]
```

```python
# Calculate and print the evaluation metrics
accuracy = metrics.accuracy_score(y_test, y_pred)
precision = metrics.precision_score(y_test, y_pred, average='weighted')
recall = metrics.recall_score(y_test, y_pred, average='weighted')
f1_score = metrics.f1_score(y_test, y_pred, average='weighted')
auc = metrics.roc_auc_score(y_test, y_pred_proba)
mcc = metrics.matthews_corrcoef(y_test, y_pred)
```

```python
print('The accuracy of the Random Forest model is :\t', accuracy)
print('The precision of the Random Forest model is :\t', precision)
print('The recall of the Random Forest model is :\t', recall)
print('The f1-score of the Random Forest model is :\t', f1_score)
print('The AUC of the Random Forest model is :\t', auc)
print('The MCC of the Random Forest model is :\t', mcc)
```

```python
importances = clf.feature_importances_
print(importances)
```
